#### Import Libraries and Dependencies 

```python
import pandas as pd
from pathlib import Path
```

#### Read in Files 

```python
# Read in data
aq1_2015 = pd.read_csv('Resources/annual_aqi_by_cbsa_2015.csv')
aq1_2016 = pd.read_csv('Resources/annual_aqi_by_cbsa_2016.csv')
aq1_2017 = pd.read_csv('Resources/annual_aqi_by_cbsa_2017.csv')
aq1_2018 = pd.read_csv('Resources/annual_aqi_by_cbsa_2018.csv')
aq1_2019 = pd.read_csv('Resources/annual_aqi_by_cbsa_2019.csv')
aq1_2020 = pd.read_csv('Resources/annual_aqi_by_cbsa_2020.csv')
```

```python
# Join by rows axis
aq1_2015_2020 = pd.concat([aq1_2015, aq1_2016, aq1_2017, aq1_2018, aq1_2019, aq1_2020], axis="rows").dropna()
aq1_2015_2020.head()
aq1_2015_2020.tail()
```

```python
#Export file
aq1_2015_2020.to_csv("Output/annual_aqi_2015_2020.csv", index=False, header=True)
```
